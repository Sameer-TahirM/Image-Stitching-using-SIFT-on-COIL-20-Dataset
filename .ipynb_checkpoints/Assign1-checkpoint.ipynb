```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

# Function to create a mask for the white duck and remove the black background
def create_duck_mask(image):
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Threshold the image to isolate the white duck
    _, mask = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)  # Adjust threshold value if needed

    # Invert the mask so that the white duck is foreground and black background is ignored
    mask = cv2.bitwise_not(mask)

    # Use the mask to isolate the duck from the background
    masked_image = cv2.bitwise_and(image, image, mask=mask)
    
    return masked_image

# Load images from the folder
image_list = []

for i in range(0, 72):
    image_list.append(cv2.imread(rf"C:\Users\samee\Desktop\University\FAST\Semester 9\CV\Assignments\Ass1\coil-20\1\obj1__{i}.png"))

# Read the first image to start stitching
img_stitched = create_duck_mask(image_list[0])

# BFMatcher object
bf = cv2.BFMatcher()

for i in range(1, 72):
    img_next = create_duck_mask(image_list[i])
    
    # Detect keypoints and descriptors using SIFT
    kp1, des1 = sift.detectAndCompute(img_stitched, None)
    kp2, des2 = sift.detectAndCompute(img_next, None)
    
    # Match descriptors with KNN
    matches = bf.knnMatch(des1, des2, k=2)

    # Apply ratio test to filter good matches
    good_matches = []
    for m in matches:
        if m[0].distance < 0.5 * m[1].distance:
            good_matches.append(m[0])
    
    matches = np.asarray(good_matches)

    # Ensure there are enough matches to compute the homography
    if len(matches) >= 4:
        src_pts = np.float32([kp1[m.queryIdx].pt for m in matches]).reshape(-1, 1, 2)
        dst_pts = np.float32([kp2[m.trainIdx].pt for m in matches]).reshape(-1, 1, 2)

        # Find the homography matrix
        H, _ = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5.0)

        # Warp the next image to the stitched result
        height, width = img_stitched.shape[:2]
        img_next_warped = cv2.warpPerspective(img_next, H, (width + img_next.shape[1], height))

        # Overlay the stitched image
        img_next_warped[0:height, 0:width] = img_stitched
        img_stitched = img_next_warped

        # Optionally visualize the result
        plt.imshow(cv2.cvtColor(img_stitched, cv2.COLOR_BGR2RGB))
        plt.title(f'Stitched Image {i}')
        plt.show()

    else:
        print(f"Not enough matches between images {i} and {i-1}")

# Save the final stitched result
cv2.imwrite('final_output.jpg', img_stitched)
```
